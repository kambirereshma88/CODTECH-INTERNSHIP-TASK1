```python
df.columns
```

# Gender Analysis

```python
ax=sns.countplot(x ='Gender',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# From above we can say  female count is more then male to purchase product 
```

```python
#here we will come to know highest amount by gender
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
```

```python
sales_genrated=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data= sales_genrated)
```

#From above graphs we can see that most of the buyers are females and 
#even the purchasing power of females are greater than man

# Now on Age Group

```python
df.columns
```

```python
ax=sns.countplot(x ='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_byAge=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data= sales_byAge,hue='Age Group')
```

## so from above graph we can see that most of the buyers are of age group between 26-35 years  that are female

# State

```python
#  to check total number of orders from top 10 states

sales_byState=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by = 'Orders',ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='State',y='Orders',data=sales_byState ,hue='State')
```

```python
#Total amount or sales from top 10 states

sales_byState=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='State',y='Amount',data=sales_byState ,hue='State')
```

# From above graph we can see that most of the orders and total sales/amount are from Uttar Pradesh,Maharashtra and Karnataka

# Marital Status

```python
ax=sns.countplot(x ='Marital_Status',data=df,hue='Marital_Status')
sns.set(rc={'figure.figsize':(15,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_byState=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.set(rc={'figure.figsize':(15,12)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_byState ,hue='Gender')
```

 From above graph we can say that most of the buyer are married(women)and they have high purchasing power

# Occupation

```python
sns.set(rc={'figure.figsize':(20,10)})
ax=sns.countplot(data=df , x ='Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_byState=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)

sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Occupation',y='Amount',data=sales_byState ,hue='Occupation')
```

From above graphs we see that most of the buyers are working in IT , AViation and Health sector


# Product Category

# Product Id
